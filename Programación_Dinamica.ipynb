```python
def cut_rod(p, n):
    r = [-1]*(n + 1)
    r[0] = 0
    s = [-1]*(n + 1)
    for i in range(1, n + 1):
        q = -1
        for j in range(1, i + 1):
            temp = p[j] + r[i - j]
            if q < temp:
                q = temp
                s[i] = j
        r[i] = q
    return r, s
n = int(input('Introduce el tama単o de la barra: '))
p = [None]
for i in range(1, n + 1):
    price = input('Introduce el valor de {} in: '.format(i))
    p.append(int(price))
r, s = cut_rod(p, n)
print('Corte optimo: ', end='')
while n > 0:
    print(s[n], end=' ')
    n -= s[n]
```

```python
def lcs(X, Y, m, n):
    if m == 0 or n == 0:
       return 0;
    elif X[m-1] == Y[n-1]:
       return 1 + lcs(X, Y, m-1, n-1);
    else:
       return max(lcs(X, Y, m, n-1), lcs(X, Y, m-1, n));
X = "CADENA"
Y = "COMUN"
print ("Tama単o de la subsecuencia comun mas larga: ", lcs(X, Y, len(X), len(Y)))
```
